```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
sns.set_style("whitegrid")
sns.set_palette("husl")
```

```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
plt.subplot(1,3,3)
sns.rugplot(data=tips,x='total_bill',hue='sex')
plt.title("Rug Plot")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15,4))
plt.subplot(1,4,1)
sns.countplot(data=tips, x='day')
plt.title('Count Plot - Days')
```

```python
plt.subplot(1,4,2)
sns.barplot(data=tips,x='day',y='total_bill',errorbar=('ci',95))
plt.title("Bar Plot - Avg Bill by Day")
```
