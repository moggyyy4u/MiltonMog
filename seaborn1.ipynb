```python
tips = sns.load_dataset("tips")
```

```python
plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
sns.histplot(data=tips, x='total_bill', bins=30, kde=True)
plt.title('Histogram with KDE')
```

```python
plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 2)
sns.kdeplot(data=tips, x='total_bill', hue='sex', shade=True)
plt.title('KDE Plot by Gender')
```

```python
plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 3)
sns.rugplot(data=tips, x='total_bill', hue='sex')
plt.title('Rug Plot')

plt.tight_layout()
plt.show()
```
